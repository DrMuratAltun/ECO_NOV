```python
#Betimsel istatistikler
ogrenci_puan=sayi_listesi=np.random.randint(1,101,100000) # büyük sayılar teoremi
```

```python
ogrenci_puan
```

```python
# ortalama
np.mean(ogrenci_puan)
```

```python
unique,counts=np.unique(ogrenci_puan, return_counts=True)
dict(zip(unique,counts))
```

```python
np.median(ogrenci_puan) #Ortanca değer
```

```python
np.std(ogrenci_puan)
```

```python
#maximum ve minumum
np.amax(ogrenci_puan), np.amin(ogrenci_puan)
```

```python
#Varyans standart sapmanın karesidir
np.var(ogrenci_puan)
```

```python
np.std(ogrenci_puan)**2
```

```python
# aykırı değerleri hesaplama
# aykırı olmayan min değer
min_aykiri_olmayan=np.mean(ogrenci_puan)-(3*np.std(ogrenci_puan))
min_aykiri_olmayan
```

```python
# aykırı olmayan max değer
max_aykiri_olmayan=np.mean(ogrenci_puan)+(3*np.std(ogrenci_puan))
max_aykiri_olmayan
```

```python
gelir=np.array([20000,30000,40000, 50000, 40000, 10000,50000, 
                40000, 50000, 40000, 10000, 50000, 40000, 50000, 40000, 
                40000, 50000, 40000, 10000, 50000, 40000, 50000, 40000, 
                10000, 10000,50000,500000])
ortalama=np.mean(gelir)
min_outlier=ortalama-(3*np.std(gelir))
max_outlier=ortalama+(3*np.std(gelir))
```

```python
min_outlier,max_outlier
```
