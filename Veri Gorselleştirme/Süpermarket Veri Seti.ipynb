```python
df_clean=df.drop(['Date','Time'],axis=1)
df_clean
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x=df_clean['month'],
             palette='Set1',)
plt.xticks(rotation=45)
plt.show()
```

```python
df_clean.groupby('month')['Total'].sum().plot(kind='bar')
```

```python
monthly_total=df_clean.groupby('month')['Total'].sum().reset_index()
monthly_total
```

```python
sns.barplot(x=monthly_total['month'],y=monthly_total['Total'])
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
sns.barplot(x='month',
            y='Total',
           data=df_clean.groupby('month')['Total'].sum().reset_index(),
           palette='Set1')
```

```python
sns.countplot(x='month',
             hue='Branch',
             data=df_clean)
```

```python
monthly_branch_total=df_clean.groupby(['month','Branch'])['Total'].sum().reset_index()
monthly_branch_total
```

```python
sns.barplot(x='month',
           y='Total',
           hue='Branch',
           data=monthly_branch_total)
```

```python
sns.lineplot(x='month',
           y='Total',
           hue='Branch',
           data=monthly_branch_total)
```
