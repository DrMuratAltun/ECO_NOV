## SeaBorn Kütüphanesi Hazır Veri Setleriyle Çalışma ## 

```python
#seaborn kutuphanesini içe aktar
import seaborn as sns 
#matplotlib kutuphanesini içe aktar
import matplotlib.pyplot as plt
```

### Veri Setini Yükle

```python
print('Kütüphanedeki veri setlerini listele')
print(sns.get_dataset_names())
```

```python
#Data Frame e yükle
bahsis_veri=sns.load_dataset('tips');
```

```python
bahsis_veri.head()
```

```python
bahsis_veri.info()
```

```python
bahsis_veri.rename(columns={'total_bill':'Toplam Tutar',
                            'tip':'Bahşiş',
                            'sex':'Cinsiyet',
                            'smoker':'Sigara',
                           'day':'Gün',
                           'time':'Zaman',
                           'size':'Kişi sayısı'},inplace=True)
```

```python
bahsis_veri.info()
```

```python
bahsis_veri
```

```python
#Cinsiyetteki değerleri değiştirelim Female:Kadın ve Male:Erkek
bahsis_veri['Cinsiyet'].replace({'Female':'Kadın','Male':'Erkek'},inplace=True)
```

```python
bahsis_veri['Cinsiyet']
```

```python
bahsis_veri['Gün'].unique()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
#Gün adlarını da Türkçe yapalım
bahsis_veri['Gün'].replace({'Sun':'Pazar', 'Sat':'Cumartesi', 'Thur':'Perşembe', 'Fri':'Cuma'},inplace=True)
```

```python
bahsis_veri['Gün'].unique()
```

```python
bahsis_veri['Sigara'].unique()
```

```python
bahsis_veri['Sigara'].replace({'Yes':'Evet', 'No':'Hayır'},inplace=True)
```

```python
bahsis_veri['Sigara']
```

```python
bahsis_veri['Zaman'].unique()
```

```python
bahsis_veri['Zaman'].replace({'Dinner':'Akşam', 'Lunch':'Öğle'},inplace=True)
```

```python
bahsis_veri['Zaman']
```

```python
bahsis_veri.info()
```

```python
bahsis_veri.describe()
```

```python
sns.scatterplot(data=bahsis_veri,
                y='Toplam Tutar',
                x='Bahşiş');
```

```python
#stil seçme
sns.set_style("dark") #whitegrid
#grafik boyuru
sns.set(rc={'figure.figsize':(15,10)})
#grafik seçmne
sns.scatterplot(data=bahsis_veri,
                y='Toplam Tutar',
                x='Bahşiş',
               hue='Cinsiyet',
               color='b',
               palette='Set1')
#Grafiğin ana başlığı
plt.title('Toplam Tutar - Bahsiş')

# y ekseni etiketi
plt.ylabel('Toplam Tutar')

# x ekseni etiketi
plt.xlabel('Bahşiş')
           
#grafiği göster
plt.show()
```

```python
#En yüksek hesap
bahsis_veri['Toplam Tutar'].max()
```

```python
#En Yüksek Bahşiş
bahsis_veri['Bahşiş'].max()
```

```python
sns.relplot(x='Toplam Tutar',
           y='Bahşiş',
           hue='Sigara',
           data=bahsis_veri,
           palette='viridis',
           height=10)
plt.title('Toplam Tutar - Bahşiş')
plt.ylabel('Bahşiş')
plt.xlabel('Toplam Tutar')
plt.show()
```

```python
#Cinsiyete göre bahşişi ortalaması
bahsis_veri.groupby(['Cinsiyet'])['Bahşiş'].mean()
```

```python
#Sigara İçme Durumuna Göre Bahşiş Ortalaması
bahsis_veri.groupby(['Sigara'])['Bahşiş'].mean()
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.relplot(x='Toplam Tutar',
            y='Bahşiş',
            size='Kişi sayısı',
            sizes=(5,200),
            hue='Cinsiyet',
            data=bahsis_veri,
           palette='Set2',
           height=10)
plt.title('Toplam Tutar - Bahşiş')
plt.ylabel('Bahşiş')
plt.xlabel('Toplam Tutar')
plt.show()
```

```python
sns.relplot(x='Toplam Tutar',
           y='Bahşiş',
           hue='Cinsiyet',
           col='Zaman',
           data=bahsis_veri,palette='Set1')
plt.show()
```

```python
# Toplam Tutar Dağılımı
sns.distplot(bahsis_veri['Toplam Tutar'], 
            bins=20)
plt.show()
```

```python
bahsis_veri['Toplam Tutar'].describe()
```

```python
sns.heatmap(bahsis_veri[['Toplam Tutar','Bahşiş','Kişi sayısı']].corr(),
            annot=True,
           cmap='coolwarm',
           linewidths=5,
           fmt='.2f')
plt.show()
```

```python
sns.displot(bahsis_veri,
           x='Bahşiş',
           hue='Cinsiyet',
           height=8)
plt.ylabel('Kişi Sayısı')
plt.show()
```

```python
bahsis_veri['Cinsiyet'].value_counts().plot.pie(autopct='%1.1f%%');
```

```python
sns.countplot(data=bahsis_veri,
              x='Cinsiyet',
              palette={'Erkek':'b','Kadın':'r'})
plt.show()
```

```python
#Günlere göre Toplam (Toplam Tutar)
bahsis_veri.groupby(['Gün'])['Toplam Tutar'].sum().plot.bar()
```

```python
#Günlere göre Ortalama Toplam Tutar
bahsis_veri.groupby(['Gün'])['Toplam Tutar'].mean().plot.bar()
```

```python
sns.countplot(data=bahsis_veri,
              x='Gün',
              palette='Set1')
plt.ylabel('Kişi Sayısı')
plt.show()
```

```python
sns.catplot(x='Gün',
           y='Toplam Tutar',
           kind='box',
           data=bahsis_veri,
           palette='Set1',
           height=10)
plt.xlabel('Günler')
plt.ylabel('Toplam Tutar')
plt.title('Toplam Tutar - Günler')
plt.show()
```

```python
sns.catplot(x='Gün',
           y='Toplam Tutar',
           hue='Cinsiyet',
           kind='violin',
           inner='stick',
            split=True,
            palette='pastel',
           data=bahsis_veri,
           height=10)
plt.xlabel('Günler')
plt.ylabel('Toplam Tutar')
plt.title('Toplam Tutar - Günler')
plt.show()
```

```python
sns.catplot(x='Sigara',
           y='Toplam Tutar',
           hue='Cinsiyet',
           kind='violin',
           inner='stick',
            split=True,
            palette='pastel',
           data=bahsis_veri,
           height=10)
plt.xlabel('Günler')
plt.ylabel('Toplam Tutar')
plt.title('Toplam Tutar - Günler')
plt.show()
```

```python
sg=sns.PairGrid(bahsis_veri)
sg.map_upper(plt.scatter)
sg.map_lower(sns.kdeplot, fill=True)
sg.map_diag(sns.histplot,kde=True)
plt.show()
```
