```python
df_select=df[['date','Turkey','Italy','Germany','China', 'US']]
```

```python
df_select.drop('date', axis=1).sum(axis=0)
```

```python
df_select.drop('date', axis=1).sum(axis=0).plot(kind='bar')
```

```python
sns.barplot(df_select.drop('date', axis=1).sum(axis=0),
           palette='rainbow')
```

```python
sum_daily=df_select.drop('date', axis=1).cumsum(axis=0)
sum_daily
```

```python
sum_daily.plot()
```

```python
sum_daily['Turkey'].plot()
```

```python
df_select['date'].dtype
```

```python
df_select['date']=pd.to_datetime(df_select['date'])
df_select['date'].dtype
```

```python
df_select.set_index('date', inplace=True)
```

```python
df_select
```

```python
df_select_daily=df_select.cumsum(axis=0)
df_select_daily
```

```python
df_select_daily['Turkey'].plot()
```
